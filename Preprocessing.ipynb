9개 다 있으면 boolean indexing으로 가져와서 각 column에 새로운 index를 추가하여 유사도 score
1. 10,11 있으면 유사도, 11,12 있으면 유사도 이런 식으로 진행
2. 연속되지 않는 것이 있다면 가져와서 보기

아니면 9개, 8개, 7개, ..., 1개로 볼까

```python
def identifying_cnt(df, crp_name):
    criteria = df[crp_name] != '0.0'
    cnt = 0
    for a in criteria:
        if a == True:
            cnt+= 1
    return cnt
```
